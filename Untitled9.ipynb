**Data Preprocessing**

Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

Pregnancies: Number of times pregnant
Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
BloodPressure: Diastolic blood pressure (mm Hg)
SkinThickness: Triceps skin fold thickness (mm)
Insulin: 2-Hour serum insulin (mu U/ml)
BMI: Body mass index (weight in kg/(height in m)^2)
DiabetesPedigreeFunction: Diabetes pedigree function
Age: Age (years)
Outcome: Class variable (0 or 1)

**Data Analysis**

In the code below, we visualize the relationship between various features and the outcome in the diabetes dataset. Each subplot represents the relationship between a specific feature (['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']) and the target variable "Outcome." We use the scatterplot function from Seaborn to create each plot, showing how each feature correlates with the likelihood of diabetes.

```python
plt.figure(figsize=(15,10))
n = 0
for feature in ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']:
  n += 1
  plt.subplot(3,3, n)
  plt.subplots_adjust(hspace = 0.5, wspace=0.5)
  sns.scatterplot(data=diabetes_dataset, x=feature, y="Outcome")

  plt.title(f"Relationship between {feature} on Outcome")
plt.show
```

**Separating features and label**

```python
x = diabetes_dataset.drop(columns="Outcome", axis=1)
y = diabetes_dataset["Outcome"]
```

**Perfrom cross validation on our dataset**

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
from statistics import mean


```

```python
model = LogisticRegression(max_iter=1000)
cv_score = cross_val_score(model,x, y, cv=5)
cv_score_mean_per = round(mean(cv_score)* 100, 2)
print(cv_score)
print(f"Accuracy score is : {cv_score_mean_per}%")
```

```python
# Fit the model to the entire dataset
model.fit(x, y)
```

**Build a predictive system**

```python
input_data = np.asarray([6,148,72,35,0,33.6,0.627,50]).reshape(1, -1)
input_predict = model.predict(input_data)
if input_predict[0] == 0:
  print("The patient does not have diabetes")
else:
  print("The patient have diabetes")



```
